Q3.Explain the use of grid-auto-row and grid-auto-column using code examples.

Ans:   In CSS Grid, grid-auto-rows and grid-auto-columns are used to define the size of implicit tracks—rows or columns that are created automatically by the grid container when there aren’t enough rows/columns defined explicitly to place all the grid items.
grid-auto-rows

  Defines the height of rows that are automatically created by the grid.

   Only applies to rows that are not explicitly declared in grid-template-rows.

 grid-auto-columns

  Defines the width of columns that are automatically created by the grid.

  Only applies to columns that are not explicitly declared in grid-template-columns.

These properties are useful when:

   Items are placed outside the explicitly defined grid.

   You're using grid-auto-flow: column or row, and new tracks need to be created to accommodate overflowing content.



```python
<style>
.container {
  display: grid;
  grid-template-columns: 100px 100px;
  grid-auto-rows: 80px;
  gap: 10px;
}

.item {
  background-color: lightcoral;
  padding: 10px;
}
</style>

<div class="container">
  <div class="item">1</div>
  <div class="item">2</div>
  <div class="item">3</div> <!-- Goes to the next row, which is created implicitly -->
  <div class="item">4</div>
</div>
```

Q5. Explain the difference between justify-items and justify-self using code examples.
Ans:
  justify-items

   Applies to: Grid containers

   Sets the alignment of items within their grid area on the inline (row) axis.

   Affects all children unless overridden.


   justify-self

  Applies to: Individual grid items

   Overrides justify-items for a specific item.

```python
<style>
.grid {
  display: grid;
  grid-template-columns: repeat(2, 100px);
  justify-items: center; /* All children centered horizontally */
  gap: 10px;
}
.box {
  background: lightblue;
}
</style>

<div class="grid">
  <div class="box">A</div>
  <div class="box">B</div>
</div>
```

```python
<style>
.grid {
  display: grid;
  grid-template-columns: repeat(2, 100px);
  justify-items: start; /* Default alignment for all */
  gap: 10px;
}
.box.special {
  justify-self: end; /* This one overrides the default */
}
.box {
  background: lightcoral;
}
</style>

<div class="grid">
  <div class="box">A</div>
  <div class="box special">B</div>
</div>
```
